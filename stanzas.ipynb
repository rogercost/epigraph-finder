```python
import requests
```

```python
# The Project Gutenberg eBook of Lyrical Ballads, With a Few Other Poems (1798)
# From the Poetry bookshelf at https://www.gutenberg.org/ebooks/bookshelf/60
# response = requests.get("https://www.gutenberg.org/cache/epub/9622/pg9622.txt")

# The Song of Hiawatha, by Longfellow
response = requests.get("https://www.gutenberg.org/cache/epub/19/pg19.txt")
```

```python
response.text
```

```python
lines = [x for x in response.text.split("\r\n")]
lines[130:140]
```

```python
len(lines)
```

```python
def segment_stanzas(lines):
  '''
  Simple rule-based segmentation, which is dependent on formatting in the source text
  '''
  stanzas = []
  current_stanza = []
  for line in lines:
    current_stanza.append(line)
    if line.strip() == '':
      if len(current_stanza[0]) > 2:
        stanzas.append("\n".join(current_stanza))
      current_stanza = []

  if current_stanza and len(current_stanza[0]) > 2:
    stanzas.append("\n".join(current_stanza))

  return stanzas
```

```python
stanzas = segment_stanzas(lines)

for stanza in stanzas[100:110]:
    print(stanza)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('Alibaba-NLP/gte-base-en-v1.5', trust_remote_code=True)
```

```python
embeddings = model.encode(stanzas[100:110])
```

```python
question_embedding = model.encode("The night sky on a clear night")
similarities = model.similarity(question_embedding, embeddings)
print(similarities)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
```

```python
hf_emb_model = HuggingFaceBgeEmbeddings(model_name="Alibaba-NLP/gte-base-en-v1.5",
                                        model_kwargs={"device": "cpu", "trust_remote_code": True},
                                        encode_kwargs={"normalize_embeddings": True})
```

```python
mini_faiss = FAISS.from_texts(stanzas[100:110], hf_emb_model)
```

```python
mini_faiss.similarity_search("The moon in a clear night sky over the ocean")
```

```python
faiss = FAISS.from_texts(stanzas, hf_emb_model)
```

```python
faiss.similarity_search("The moon in a clear night sky over the ocean")
```

```python
# Now, download a book and generate epigraphs for its sections.

book_response = requests.get("https://www.gutenberg.org/cache/epub/73987/pg73987.txt")
delimiter = "*       *       *       *       *"

book_sections = book_response.text.split(delimiter)
len(book_sections)
```

```python
print(book_sections[1])
faiss.similarity_search(book_sections[1])
```

```python
for section in book_sections:
  print(f"{section[:100]} => {faiss.similarity_search(section)[0].page_content}")
```

```python
results = faiss.similarity_search("""

Whenever I have done right, it has always seemed to me that somebody or
something helped me. But when I have gone wrong, I have sinned through
no one’s fault but my own. No man ever made me do wrong. The man who
has the headache next morning is the fellow who transgressed the night
before. The sinner can no more shift his responsibility than he can
wish his headache off on the other fellow.

""")

for result in results:
  print(result.page_content)
```
